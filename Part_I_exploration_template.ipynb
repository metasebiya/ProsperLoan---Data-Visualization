## Bivariate Exploration

> In this section I will check the pairwise correlation among features.

```python
#Separately holding numerical and categorical values.
numeric_vars = ['StatedMonthlyIncome', 'LoanOriginalAmount', 'MonthlyLoanPayment', 'BorrowerAPR', 'BorrowerRate','ProsperScore','CreditScoreRangeUpper']
categoric_vars = ['LoanStatus', 'EmploymentStatus','IncomeRange','Term']
```

### Checking the relationship btween numerical variables

### Question
#### How are the numeric variables related among each other?

```python
# correlation plot
plt.figure(figsize = [8, 5])
sb.heatmap(prosper_df[numeric_vars].corr(), annot = True, fmt = '.3f',
           cmap = 'vlag_r', center = 0)
x_y_t('','','Relationship between numeric variables')
```

### Observation
#### From the heatmap, we can conclude that some variables have a high correlation with other variables. On the other hand, Some of the relationship have a correlation value near to 0. Lastly, I have found one relationship that has a correlation value that draws to 1.

### Relationship between LoanOriginalAmount and BorrowerAPR variables

### Observation
